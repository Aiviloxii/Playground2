# Data Visualization
- Visual Representation of data to discover trends
- Helps in story-telling and decision-making

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all plots aka subplots
- Axes: This is a single plot

```python
# Create a figure with a single axes
x = [1, 2, 3]

y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show() #to display the plot
```

```python
# Create a figure with multiple subplots

fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title("Left Plot")
axes[1].plot(x,y)
axes[1].set_title("Middle Plot")


plt.show()
```

## EDA: Exploratory Data Analysis
- Either Univariate or Multivariate
- Univariate analysis means examining one variable at a time

```python
df["Ship Mode"].value_counts()
```

```python
df["Region"].value_counts()
```

What are the modes of shipping used mostly in a region?

```python
region_mode_of_shipping = df.groupby("Ship Mode")["Region"].count().reset_index()

region_mode_of_shipping
```

```python
plt.plot(region_mode_of_shipping["Ship Mode"], region_mode_of_shipping["Region"], color='grey', marker='*')
plt.title("Different shipping modes in different regions")
plt.xlabel("Ship Modes", color='red')
plt.ylabel("Regions", color='red')
plt.show()
```
