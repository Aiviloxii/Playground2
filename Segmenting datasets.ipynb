```python
total_sales_California = data[data["State"] == "California"]["Sales"].sum()
```

```python
total_sales_Ohio = data[data["State"] == "Ohio"]["Sales"].sum()
```

```python
total_sales_Ohio
```

```python
total_sales_by_state = data.groupby("Stateto")["Sales"].sum().reset_index()

total_sales_by_state
```

```python
total_sales_by_state.sort_values(by='Sales', ascending=False, inplace=True)
```

```python
total_sales_by_state
```

```python
avg_sales_by_state = data.groupby("State")["Sales"].mean().reset_index()
avg_sales_by_state.sort_values(by="Sales", ascending=False, inplace=True)

avg_sales_by_state
```

```python
avg_total_sales = data.groupby("State")["Sales"].agg([np.sum, np.mean, np.min, np.max]).reset_index()

avg_total_sales
```

```python
data.head()
```

```python
data["City"].value_counts()
```

```python
avg_sales_by_city = data.groupby("City")["Sales"].mean().reset_index()

avg_sales_by_city.tail()
```

```python
avg_sales_by_category = data.groupby(["Category", "Sub-Category"])["Sales"].mean().reset_index()

avg_sales_by_category
```

```python
sales_by_sub_category = data.groupby(["Category", "Sub-Category"])["Sales"].agg([np.sum, np.mean, np.min, np.max])
```

```python
sales_by_sub_category
```

## Pivoting

```python
sales_by_state = data.pivot_table(index="Segment", values="Sales", columns="Sub-Category", aggfunc=np.sum, fill_value=0)

sales_by_state
```
