```python
import pandas as pd

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv("../Datasets/pizzas.csv")

df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv("../Datasets/pizza_types.csv", encoding='latin1')

df2.head()
```

```python
df2.info()
```

```python
df3 = pd.read_csv("../Datasets/orders.csv")

df3.head()
```

```python
df3.info()
```

```python
df3["date"] = pd.to_datetime(df3["date"])
```

```python
df3.info()
```

```python
df4 = pd.read_csv("../Datasets/order_details.csv")

df4.head()
```

```python
df4.info()
```

```python
pizza_data = pd.merge(df1, df2, on="pizza_type_id", how="outer")

pizza_data
```

```python
order_data = pd.merge(df3, df4, on="order_id", how='outer')

order_data
```

```python
pizza_data.info()
```

```python
order_data.info()
```

```python
pizza_sales_data = pd.merge(pizza_data, order_data, on='pizza_id', how='inner')

pizza_sales_data 
```

```python
pizza_sales_data.info()
```

```python
pizza_sales_data.head(10)
```

### Total Revenue is the summation of all prices and quantity
- first find the revenue
- then sum it to get Total Revenue/Sales

```python
pizza_sales_data['revenue'] = (pizza_sales_data['price'] * pizza_sales_data['quantity'])
```

```python
pizza_sales_data.head()
```

```python
total_revenue = pizza_sales_data['revenue'].sum()
total_revenue  # this is the total sales for the year
```

#### To find Total Quantity:
- sum all values in the quantity column

```python
pizza_sales_data['quantity'].sum() # this is total number of pizzas sold in the year
```

#### To find Total Orders:
- select order ids and count

```python
pizza_sales_data['order_id'].count() #this is the total orders made to the pizza place in the year
```

#### To find pizza types they sell:
- count pizza names and type ids

```python
pizza_types = pizza_sales_data.groupby("name")['pizza_type_id'].nunique().reset_index()

pizza_types #this shows the types of pizza they sell
```

#### To find the average price of pizza sold:
- calculate the mean of the prices 

```python
pizza_sales_data['price'].mean() # this is the average price of pizzas sold
```

#### To find the Peak hours of Sales:
- using the time column

```python
print(pizza_sales_data['time'].head(10))
```

```python
pizza_sales_data['time'] = pd.to_datetime(pizza_sales_data['time'], errors='coerce').dt.time # convert time column to time data type
```

```python
print(pizza_sales_data['time'].head(10))
```

```python
pizza_sales_data['sale hours'] = pizza_sales_data['time'].apply(lambda x:x.hour) #extract the hours
```

```python
pizza_sales_data.head()
```

```python
peak_sales = pizza_sales_data.groupby("sale hours")['order_id'].count().reset_index()

# get total orders at a particular hour
```

```python
peak_sales.sort_values(by='order_id', ascending=False) # sort in descending order
```

##### below shows that the peak hours of sales are: 12pm and 1pm, followed by 6pm

```python
plt.bar(peak_sales['sale hours'], peak_sales['order_id'], color='lightgreen')
plt.title("Peak hours of Sales", color='green')
plt.xlabel(" Hour of the Day", color='green')
plt.ylabel("Number of Orders", color='green')

plt.show()
# visual representation
```

#### To find total sales made on each day:
- using the date column to find most orders

```python
pizza_sales_data['day_of_sales'] = pizza_sales_data['date'].dt.day_name()
```

```python
pizza_sales_data
```

```python
total_sales_per_week = pizza_sales_data.groupby("day_of_sales")['revenue'].sum().reset_index()

total_sales_per_week #total sales made on each day
```

```python
total_sales_per_week.sort_values(by='revenue', ascending=False)
```

##### shows the most sales is on Friday

```python
plt.bar(total_sales_per_week['day_of_sales'], total_sales_per_week['revenue'], color='cyan')
plt.title("Total Sales per Day", color='navy', fontsize=20)
plt.xlabel("Days of the Week", color='navy')
plt.xticks(rotation=90)
plt.ylabel("Revenue made each day", color='navy')

plt.show() 
#visual representation of sales per day
```

### To find 5 best selling pizzas:
- get quantity of each type of pizza sold

```python
bestselling_pizza = pizza_sales_data.groupby('name')['quantity'].sum().reset_index()

bestselling_pizza
```

```python
bsp = bestselling_pizza.sort_values(by='quantity', ascending=False).head(10)
```

```python
bsp
```

```python
plt.bar(bsp['name'], bsp['quantity'], color='orange')
plt.title("Top 5 Best Selling Pizza", fontsize=20, color='brown')
plt.xlabel("Pizza Types", color='brown')
plt.xticks(rotation=90)
plt.ylabel("Quantity Sold", color='brown')

plt.show()
```

#### TO get sales made each month and any noticeable trend:
- get the sales made per month from date column

```python
pizza_sales_data['month of sale'] = pizza_sales_data['date'].dt.month_name()
```

```python
pizza_sales_data.head()
```

```python
monthly_sales = pizza_sales_data.groupby('month of sale')['revenue'].sum().reset_index()

monthly_sales
```

```python
monthly_sales.sort_values(by='revenue', ascending=False)
```

```python
plt.plot(monthly_sales['month of sale'], monthly_sales['revenue'], color='blue', marker='o')
plt.title("Monthly Sales", fontsize=20, color='navy')
plt.xlabel("MONTHS", color='navy')
plt.xticks(rotation=90)
plt.ylabel("Revenue made each Month", color='navy')

plt.show()
# visual representation 
```

#### There is a noticeable trend of Pizza bought less in cold seasons and more in warmer seasons

#### To find underperforming Pizzas:
- get from quantity and pizza name

```python
underperformers = pizza_sales_data.groupby('name')['quantity'].sum().reset_index()

underperformers
```

```python
underperformers.sort_values(by='quantity')
```

##### With Brie Carrie Pizza being the least sold, it is not doing so well

```python
plt.bar(underperformers['name'], underperformers['quantity'], color='grey')
plt.title("Least Sold Pizzas", fontsize=20)
plt.xlabel("Pizza Types", color='red')
plt.xticks(rotation=90)
plt.ylabel("Quantity", color='red')

plt.show()
```
